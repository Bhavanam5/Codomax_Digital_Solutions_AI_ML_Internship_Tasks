```python
import pandas as pd

# Create Student Dataset
data = {
    "Student": ["Raha", "Ayra", "Priya", "Ryan", "Neha", "Kiran", "Arun", "Vijay"],
    "Hours_Studied": [2, 5, 4, 5, 6, 3, 4, 5],
    "Attendance": [77, 90, 80, 95, 98, 72, 84, 95],
    "Previous_Marks": [70, 75, 55, 90, 67, 68, 78, 85],
    "Final_Score": [75, 90, 80, 95, 92, 88, 82, 97]
}

# Create DataFrame
df = pd.DataFrame(data)

print("Student Dataset")
print(df)

print("\nFirst Five Records")
print(df.head())

print("\nLast Five Records")
print(df.tail())

print("\nDataset Information")
print(df.info())

print("\nStatistical Summary")
print(df.describe())

print("\nStudent Names")
print(df["Student"])

print("\nHours Studied and Final Score")
print(df[["Hours_Studied", "Final_Score"]])

print("\nStudents who studied more than 5 hours")
print(df[df["Hours_Studied"] > 5])

print("\nAverage Final Score:", df["Final_Score"].mean())
print("Highest Final Score:", df["Final_Score"].max())
print("Lowest Final Score:", df["Final_Score"].min())

print("\nStudents sorted by Final Score")
print(df.sort_values(by="Final_Score", ascending=False))
```
